```python
import nltk
from nltk.translate.bleu_score import sentence_bleu
import pandas as pd

# Perguntas originais (manualmente traduzidas para o inglês)
original_translations = [
    "How many patients were admitted in the last month?",
    "Which patients were admitted more than once in the last year?",
    "How many admissions had complications during treatment?",
    "Which admissions resulted in death in the last quarter?",
    "What is the bed occupancy rate at 'São José' hospital?",
    "What are the most common ICDs among patients admitted to the Cardiology department?",
    "How many patients under 12 years old were admitted to the Pediatrics department?",
    "What is the average length of stay per patient?",
    "What are the main reasons for admission at Luz hospital?",
    "Which patients were discharged after improvement following antibiotic treatment?",
    "What is the hospital readmission rate within 30 days after discharge?",
    "What were the top 5 ICDs for admissions in the last semester?",
    "What are the 3 most prescribed medications for patients with heart failure?",
    "How many imaging exams were performed in the last 6 months?",
    "Which exams resulted in a conduct to be evaluated?",
    "What is the number of patients admitted through emergency and scheduled consultations?",
    "Who are the 10 patients with the most admissions in the last year?",
    "Which doctors performed the most treatments this month?",
    "Which admissions occurred at the 'Universitário de Brasília' hospital?",
    "How many patients were discharged with complications during hospitalization?",
    "How many pediatric beds are currently occupied?",
    "Which patients have a history of hypertension and were admitted with complications?",
    "What treatments were applied to patients with diabetes mellitus?",
    "Which hospitals had the most admissions in the Emergency department?",
    "Which admissions resulted in transfers to another hospital in the last quarter?"
]

# Perguntas traduzidas pelo Libre Translate
libre_translations = [
    "How many patients were admitted in the last month?",
    "What patients were hospitalized more than once in the last year?",
    "How many hospitals had complications during treatment?",
    "What hospitalizations resulted in death in the last quarter?",
    "What is the bed occupancy index in the 'Saint Joseph' hospital?",
    "What are the most common ICDs among patients admitted to the Department of Cardiology?",
    "How many patients under 12 were admitted to the Department of Pediatrics?",
    "What is the average of days of hospitalization per patient?",
    "What are the main reasons for hospitalization in the Luz hospital?",
    "What patients have had high for improvement after an antibiotic treatment?",
    "What is the hospital readmission rate within 30 days after discharge?",
    "What were the top 5 ICDs of admissions in the last semester?",
    "What are the 3 most prescribed medicines for patients with heart failure?",
    "How many imaging tests have been performed in the last 6 months?",
    "What tests resulted in a conduct being evaluated?",
    "What is the number of patients admitted by emergency and by scheduled appointments?",
    "Who are the 10 most hospitalized patients in the last year?",
    "What doctors have performed more treatments this month?",
    "What were the hospital admissions in the hospital 'Universitário de Brasília'?",
    "How many patients were discharged with complications during hospitalization?",
    "How many pediatric beds are currently occupied?",
    "What patients have history of hypertension and have been hospitalized with complications?",
    "What treatments have been applied to patients with diabetes mellitus?",
    "What hospitals have had more hospitalizations in the Emergency Department?",
    "What hospitals resulted in transfers to another hospital in the last quarter?"
]

# Cálculo do BLEU Score usando n-gramas para cada tradução
bleu_scores_libre = [
    sentence_bleu([ref.split()], trans.split()) for ref, trans in zip(original_translations, libre_translations)
]

# Construção do DataFrame para exibir os resultados
df_bleu_libre = pd.DataFrame({
    "Original Question (EN)": original_translations,
    "Libre Translation": libre_translations,
    "BLEU Score": bleu_scores_libre
})

# Ajustar a exibição do pandas para evitar notação científica
pd.set_option('display.float_format', '{:.6f}'.format)

# Exibir o DataFrame novamente
df_bleu_libre
```
